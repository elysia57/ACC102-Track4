```python
## 1. Import Required Libraries
# This section imports essential packages for data manipulation and visualization.
# Pandas is used for data cleaning, calculation and table management.
# Matplotlib is used to generate static trend charts for financial comparison.

import pandas as pd
import matplotlib.pyplot as plt
```

```python
## 2. Construct Raw Dataset
# Manually input official public financial data from annual reports.
# Key variables include annual revenue, net income and total equity.
# Data covers five consecutive years for two listed technology companies.

data = {
    "tic": ["AAPL"] * 5 + ["MSFT"] * 5,
    "year": [2020, 2021, 2022, 2023, 2024] * 2,
    "revenue": [274515, 365817, 394328, 383285, 381652,
                143015, 168088, 198270, 211915, 245123],
    "net_income": [57413, 94680, 99803, 96995, 112238,
                   44281, 72738, 72361, 78032, 88136],
    "total_equity": [65337, 71932, 50672, 62122, 73860,
                     119462, 141988, 166542, 205678, 234686]
}

# Convert dictionary structure into pandas DataFrame for further analysis
df = pd.DataFrame(data)
```

```python
## 3. Basic Data Inspection
# Check data structure, column definition and overall data completeness.
# Ensure no missing values and correct grouping by company and year.

print("=== Dataset Overview ===")
print(df.head(10))
```

```python
## 4. Calculate Core Financial Ratios
# Based on raw financial data, compute two essential profitability indicators.
# ROE = net income / total equity, reflecting shareholder capital efficiency.
# Profit margin = net income / revenue, reflecting overall operating profitability.

df["ROE"] = df["net_income"] / df["total_equity"]
df["profit_margin"] = df["net_income"] / df["revenue"]
```

```python
## 5. Display Processed Complete Data
# Output the full table containing original data and newly calculated ratios.
# Facilitate data checking and support subsequent visual analysis.

print("\n=== Complete Data with Calculated Financial Ratios ===")
print(df)
```

```python
## 6. Descriptive Statistical Analysis
# Group data by company to compare average performance over five years.
# Summarise the overall level of profitability and operating scale.

print("\n=== Average Financial Performance by Company ===")
summary = df.groupby("tic")[["ROE", "profit_margin", "revenue"]].mean()
print(summary)
```

```python
## 7. Data Filtering for Independent Comparison
# Separate the dataset into two individual company subsets.
# Prepare independent data for trend drawing and horizontal comparison.

df_apple = df[df["tic"] == "AAPL"]
df_microsoft = df[df["tic"] == "MSFT"]
```

```python
## 8. Visual Trend Visualization
# Plot annual ROE trend for both companies from 2020 to 2024.
# Visually reflect the changing trend and performance gap.

plt.figure(figsize=(10, 5))
plt.plot(df_apple["year"], df_apple["ROE"], marker = "o", label = "Apple")
plt.plot(df_microsoft["year"], df_microsoft["ROE"], marker = "s", label = "Microsoft")

plt.title("ROE Trend Comparison: Apple vs Microsoft (2020–2024)")
plt.xlabel("Year")
plt.ylabel("Return on Equity")
plt.legend()
plt.grid(alpha = 0.3)
plt.show()
```

```python
## 9. Overall Analysis and Key Findings
# Conclude objective financial performance differences between the two firms.
# Combine data and charts to complete logical analytical reasoning.

print("\n=== Key Analytical Findings ===")
print("1. Apple maintains a higher and more stable ROE level than Microsoft.")
print("2. Both companies present stable profitability and positive net income.")
print("3. Continuous revenue growth proves sustainable business expansion.")
print("4. Calculated financial ratios effectively reflect operational quality.")
```
